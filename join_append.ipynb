```python
import pandas as pd
```

```python
df1=pd.DataFrame({'A':[1,2,3,4],
                'Class':[9,10,11,12]})
df2=pd.DataFrame({'B':[1,2,3,4],
                'Name':['A','B','C','D']})
display(df1,df2)
```

```python
df1.join(df2)
```

```python
df1=pd.DataFrame({'A':[1,2,3,4],
                'Class':[9,10,11,12]})
df2=pd.DataFrame({'B':[1,2,3],
                'Name':['A','B','C']})
display(df1,df2)
```

```python
df1.join(df2)
```

```python
df1.join(df2,how='inner')        # Intersection 
```

```python
df1.join(df2,how='outer')          # outer means union
```

```python
df1.join(df2,how='left')
```

```python
df1.join(df2,how='right')
```

```python
df1=pd.DataFrame({'A':[1,2,3,4],
                'Class':[9,10,11,12]})
df2=pd.DataFrame({'A':[1,2,3],
                'Name':['A','B','C']})
display(df1,df2)
```

```python
df1.join(df2,lsuffix='NO1', rsuffix='No2')
```

```python
df1.join(df2,lsuffix='NO1')
```

```python
df1.join(df2,rsuffix='NO1')
```
