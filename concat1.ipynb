```python
import pandas as pd
```

```python
df1=pd.DataFrame({'a':11,'b':22,'c':33},
                 {'a':12,'b':23,'c':34},
                 {'a':13,'b':24,'c':34})
df1
```

```python
df2=pd.DataFrame({'a':12,'b':24,'c':23},
                 {'a':13,'b':25,'c':94},
                 {'a':14,'b':27,'c':30})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=0)
```

```python
pd.concat([df1,df2],axis=0,ignore_index=True)
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df2],join='inner')
```

```python
pd.concat([df1,df2],join='outer')
```

```python
pd.concat([df1,df2],keys=['df1','df2'])
```

```python
pd.concat([df1,df2],keys=['First DF','Second DF'],axis=1)
```
