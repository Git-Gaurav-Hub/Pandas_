```python
import pandas as pd
```

```python
df=pd.DataFrame({'ID':[5,6,6,8],
                'Name':[11,22,33,44],
                'rollno':[1,2,2,4],
                'year':[2009,2010,2011,2012]})
df
```

```python
df.pivot(index='year',columns='rollno')
```

```python
df.pivot_table(index='year',columns='rollno',aggfunc=['mean','sum'],margins=True)
```

```python
df.pivot_table(index='year',columns='rollno',aggfunc=['mean'])
```

```python
df.pivot_table(index='year',columns='rollno',aggfunc=['mean'],margins=True)
```

```python
df.pivot_table(index='year',columns='rollno',aggfunc=['mean'],margins=True,fill_value=0)
```
